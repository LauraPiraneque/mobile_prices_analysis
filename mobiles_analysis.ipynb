```python
mobile_dataset.drop_duplicates(inplace=True)
```

```python
mobile_dataset.shape
```

```python
currency_abbreviation = [abbr[0:3] for abbr in mobile_dataset.iloc[0, 9:14]]
currency_abbreviation
```

```python
for col in mobile_dataset.columns[9:14]:
    mobile_dataset[col] = mobile_dataset[col].str.extract(
        '([0-9]+[,./]*[0-9]*)').replace(',', '', regex=True).astype(float, errors='ignore')

mobile_dataset.head()
```

```python
dollar_convertion = [(9, 0.0036), (10, 0.012), (11, 0.14), (13, 0.27)]

for col, conv in dollar_convertion:
    column_name = mobile_dataset.columns[col]
    mobile_dataset[f'{column_name} to USD'] = mobile_dataset[column_name].apply(
        lambda x: x*conv)

mobile_dataset.head()
```

```python
mobile_dataset.isnull().sum()
```

```python
mobile_dataset.dropna(subset=['Launched Price (Pakistan)'], inplace=True)
```

```python
plt.figure(figsize=(15, 5))

mobile_dataset.groupby('Company Name')['Model Name'].count(
).sort_values(ascending=False).plot(kind='bar')

plt.title('Number of Mobile Models per Company')
plt.xlabel('Company Name')
plt.ylabel('Mobile Models')

plt.show()
```

```python
gs = gridspec.GridSpec(5, 1)
fig = plt.figure(figsize=(10, 30))

for col in np.arange(0, 5):
    column_name = mobile_dataset.columns[col + 9]
    ax = fig.add_subplot(gs[col, 0])
    ax.boxplot(mobile_dataset[column_name], vert=False)
    ax.set_title(f'{column_name} distribution')
    ax.set_xlabel(f'Price ({currency_abbreviation[col]})')


plt.show()
```

```python
plt.boxplot(mobile_dataset.loc[:, ['Launched Price (Pakistan) to USD', 'Launched Price (India) to USD',
            'Launched Price (China) to USD', 'Launched Price (USA)', 'Launched Price (Dubai) to USD']], tick_labels=currency_abbreviation)

plt.title('Price Distribution per Country')
plt.xlabel('Country')
plt.ylabel('USD')

plt.show()
```
