```python
mobile_dataset[mobile_dataset.duplicated(keep=False)]
```

```python
mobile_dataset.drop_duplicates(inplace=True)
```

```python
mobile_dataset.shape
```

```python
currency_abbreviation = [abbr[0:3] for abbr in mobile_dataset.iloc[0, 9:14]]
currency_abbreviation
```

```python
for col in mobile_dataset.columns[9:14]:
    mobile_dataset[col] = mobile_dataset[col].str.extract(
        '([0-9]+[,./]*[0-9]*)').replace(',', '', regex=True).astype(float, errors='ignore')

mobile_dataset.head()
```

```python
mobile_dataset.isnull().sum()
```

```python
mobile_dataset.dropna(subset=['Launched Price (Pakistan)'], inplace=True)
```

```python
mobile_dataset.isnull().sum()
```

```python
mobile_dataset.info()
```

```python
mobile_dataset.describe()
```

```python
plt.figure(figsize=(15, 5))

mobile_dataset.groupby('Company Name')['Model Name'].count(
).sort_values(ascending=False).plot(kind='bar')

plt.title('Number of Mobile Models per Company')
plt.xlabel('Company Name')
plt.ylabel('Mobile Models')

plt.show()
```
